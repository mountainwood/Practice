### 範例1：博客來書店搜尋"Python"

```python
import requests
res = requests.get("http://search.books.com.tw/search/query/key/python/cat/all")
print (res.text)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(res.text,'html.parser')
print (soup.title.string)
```

```python
import pandas as pd
#爬取書名
books = pd.Series()
for book in soup.select("img[class='itemcov']"):
    books = books.append(pd.Series([book['alt']])).reset_index(drop=True)  #加到pd.Series

#爬取定價
i = 0
prices = pd.Series()
for tag in soup.select('b'):
    if i%2==1 and i/2<books.size: #會抓到打折數和價格，只取價格
        prices = prices.append(pd.Series([tag.string])).reset_index(drop=True) #加到pd.Series
    i+=1
```

```python
#合併成DataFrame
df = pd.DataFrame({'書名':books, '價格': prices})
df[['書名','價格']]
```

## 4. JSON 

```python
import json
import requests
res = requests.get("http://cloud.culture.tw/frontsite/trans/SearchShowAction.do?method=doFindTypeJ&category=6")
print(res.text)
```

```python
result = json.loads(res.text) #經過json.load後，從string轉為list type
result[0:10]
```

```python
import pandas as pd
df2 = pd.DataFrame(result)
df2
```

```python
#內層的showInfo還是JSON格式
pd.DataFrame(df2['showInfo'][:10])
```

```python
# 注意階層式的資料，如showInfo這一層裡的資料還是JSON格式（對Python而言df2['showInfo']是Series,df2['showInfo'][0]是list）
print (type(df2['showInfo']))
print (type(df2['showInfo'][0]))
df3 = pd.DataFrame(df2['showInfo'][0]) 
df3
```

```python
# 全部showInfo放進DataFrame
df4 = pd.DataFrame()
for item in df2['showInfo']:
    df4 = df4.append(pd.DataFrame(item))

df4
```

```python
# 可以加上上一層的資料，如title
df = pd.DataFrame(result[0]['showInfo'])
df['title'] = result[0]['title']
df
```

## 5. XML

```python
import xml.etree.ElementTree as et
import requests
res = requests.get("http://cloud.culture.tw/frontsite/trans/emapOpenDataAction.do?method=exportEmapXMLByMainType&mainType=10")
print(res.text)
```

```python
#如果要從XML file匯入
#tree = et.ElementTree("file name") 
#root = tree.getroot()

#從XML string匯入
tree = et.ElementTree(et.fromstring(res.text))
root = tree.getroot()
root.tag, root.attrib
```

```python
# 掃描xml tree裡的所有tag內容
for elem in tree.iter():
    print (elem.tag, elem.attrib)
```

```python
# 只取得指定tag資料
info = []
for elem in tree.iter(tag='Info'):
    print (elem.tag, elem.attrib)
    info.append(elem.attrib)
```

```python
import pandas as pd
print(type(info))
df_info = pd.DataFrame(info)
```
